# Analyzing Text about Greek Mythology

```python
url = 'https://en.wikipedia.org/wiki/Greek_mythology'
```

```python
#Getting the data
import requests 

text = requests.get(url).content.decode('utf-8')
print(text[:1000])
```

```python
#Converting the data to a form that its ready for data processing
#Using built-in HTMLParser object from Python
from html.parser import HTMLParser

#Subclass the HTMLParser class and define the code that will collect all text inside HTML tags, except <script> and <style> tags.
class MyHTMLParser(HTMLParser):
    script=False
    res=""
    def handle_starttag(self, tag, attrs):
        if tag.lower() in ["script","style"]:
            self.script = True
    def handle_endtag(self, tag):
        if tag.lower() in ["script","style"]:
            self.script = False
    def handle_data(self, data):
        if str.strip(data)=="" or self.script:
            return
        self.res += ' '+data.replace('[ edit ]','')

parser = MyHTMLParser()
parser.feed(text)
text = parser.res
print(text[:1000])
```

```python
#Getting insights
#Extract keywords from the text
#Rake libraray is used for keyword extraction
import sys
!{sys.executable} -m pip install nlp_rake 
#installing library rake
```

```python
import nlp_rake
extractor=nlp_rake.Rake(max_words=2,min_freq=4,min_chars=6)
res=extractor.apply(text)
res
```

```python
#visualisations
import matplotlib.pyplot as plt
def plot(pair_list):
    k,v = zip(*pair_list)
    plt.bar(range(len(k)),v)
    plt.xticks(range(len(k)),k,rotation='vertical')
    plt.rcParams["figure.figsize"]=20,20
    plt.show()

plot(res)
```

```python
def plot(pair_list):
    k,v = zip(*pair_list)
    plt.plot(range(len(k)),v,"r--",color='g')
    plt.xticks(range(len(k)),k,rotation='vertical')
    plt.rcParams["figure.figsize"]=20,20
    plt.show()

plot(res)
```
